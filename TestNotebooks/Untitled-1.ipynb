```python
import matplotlib.pyplot as plt
import numpy as np

# Parameters
short_tax = 0.35
long_tax = 0.225
Starting_val = 70000
Current_val = 117000
Non_Quantum_stock_profit = 1000 + 1000 + 3000 + 8500 + 2000 + 1000 + 1500
print(f"Non-Quantum stock profit: ${Non_Quantum_stock_profit} Before Tax, {Non_Quantum_stock_profit * (1 - long_tax)} After Tax")

Expected_Loss = np.arange(0, 30000, 500)

# Profit calculations
keep_profit = ((Current_val - Starting_val - Non_Quantum_stock_profit - Expected_Loss) * (1 - long_tax)) + Starting_val
sell_profit = np.full_like(Expected_Loss, (Current_val - Starting_val - Non_Quantum_stock_profit ) * (1 - short_tax) + Starting_val)

# Find intersection (where difference changes sign)
diff = keep_profit - sell_profit
sign_changes = np.where(np.diff(np.sign(diff)))[0]

if len(sign_changes) > 0:
    i = sign_changes[0]
    # Linear interpolation for smoother intersection estimate
    x1, x2 = Expected_Loss[i], Expected_Loss[i + 1]
    y1, y2 = diff[i], diff[i + 1]
    intersection_loss = x1 - y1 * (x2 - x1) / (y2 - y1)
    intersection_profit = ((Current_val - Starting_val - Non_Quantum_stock_profit - intersection_loss) * (1 - long_tax)) + Starting_val
    print(f"Intersection (breakeven) at Expected Loss ≈ ${intersection_loss:,.2f}, or {Current_val - intersection_loss:,.2f} dollars left")
    print(f"Profit at intersection ≈ ${(intersection_profit - Starting_val):,.2f}")
else:
    print("No intersection found (one option is always better).")

# Plot
plt.figure(figsize=(8, 5))
plt.plot(Expected_Loss, sell_profit, 'r--', label='Sell Now (Short-Term Tax)')
plt.plot(Expected_Loss, keep_profit, 'b-', label='Keep (Long-Term Tax)')

if len(sign_changes) > 0:
    plt.scatter(intersection_loss, intersection_profit, color='k', zorder=5)
    plt.text(intersection_loss, intersection_profit,
             f'  Breakeven ≈ ${intersection_loss:,.0f}',
             verticalalignment='bottom', fontsize=9)

plt.title('Sell vs Keep Profit over Expected Loss')
plt.xlabel('Expected Loss ($)')
plt.ylabel('After-Tax Profit ($)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Parameters
short_tax = 0.35
long_tax = 0.225

Current_val = 126000
Starting_val = 70000
daily_loss = 2000     # expected loss per day
days = 10             # number of days to simulate

# Track daily values
values = [Current_val]
for _ in range(1, days):
    values.append(values[-1] - daily_loss)

# Compute after-tax profits
keep_profit = [((v - Starting_val) * (1 - long_tax) + Starting_val) for v in values]
sell_profit = [(Current_val - Starting_val) * (1 - short_tax)] * days 
sell_profit = [x + Starting_val for x in sell_profit]

# Plot results
plt.figure(figsize=(8, 5))
plt.plot(range(1, days + 1), sell_profit, 'r--', label='Sell Now (Short-Term Tax)')
plt.plot(range(1, days + 1), keep_profit, 'b-', label='Keep (Long-Term Tax, Declining Value)')

plt.title('Sell vs Keep Profit Over Time with Daily Expected Loss')
plt.xlabel('Day')
plt.ylabel('After-Tax Profit ($)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
